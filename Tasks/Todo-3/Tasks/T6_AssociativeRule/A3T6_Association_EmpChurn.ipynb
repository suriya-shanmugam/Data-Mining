```python
import pandas as pd

train_data = pd.read_csv(directory+'HR-Employee-Attrition.csv')
train_data.head()
```

```python
train_data.drop(columns=['BusinessTravel','DailyRate','Education','EducationField','EmployeeCount','EmployeeNumber','RelationshipSatisfaction'], axis=1)
```

```python
train_data = train_data.drop(columns=['BusinessTravel','DailyRate','Education','EducationField','EmployeeCount','EmployeeNumber','RelationshipSatisfaction'], axis=1)
train_data.head()
```

```python
train_data = train_data.drop(columns=['EnvironmentSatisfaction','HourlyRate','Gender','JobInvolvement','JobRole','PerformanceRating','StandardHours'], axis=1)
train_data.head()
```

```python
train_data = train_data.drop(columns=['JobLevel','JobSatisfaction','MaritalStatus','NumCompaniesWorked','OverTime','PercentSalaryHike','StockOptionLevel'], axis=1)
train_data.head()
```

```python
train_data = train_data.drop(columns=['Over18','TotalWorkingYears','TrainingTimesLastYear','YearsAtCompany','WorkLifeBalance','YearsSinceLastPromotion','YearsWithCurrManager'], axis=1)
train_data.head()
```

```python
train_data = train_data.drop(columns=['MonthlyRate','MonthlyIncome'], axis=1)
train_data.head()
```

```python
basket = pd.get_dummies(train_data[['Attrition','Department']], prefix=['Att','Dept'])
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
frequent_itemsets = apriori(basket, min_support=0.1, use_colnames=True)
print("Frequent Itemsets:")
print(frequent_itemsets)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.barplot(x='support', y='itemsets', data=frequent_itemsets, palette='viridis')
plt.title('Frequent Itemsets')
plt.xlabel('Support')
plt.ylabel('Itemsets')
plt.xticks(rotation=45)
plt.show()
```
