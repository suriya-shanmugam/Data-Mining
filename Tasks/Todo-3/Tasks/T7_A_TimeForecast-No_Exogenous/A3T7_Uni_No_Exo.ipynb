```python
import pandas as pd
train_data = pd.read_csv(directory+'champagne.csv')
train_data.head()
```

```python
train_data.plot()
```

```python
train_data['Month'] = pd.to_datetime(train_data['Month'], format='%Y-%m')
```

```python
train_data.head()
```

```python
train_data.set_index('Month', inplace=True)
```

```python
train_data.head()
```

```python
train_data.plot()
```

```python
import time
import numpy as np
import pandas as pd


from pycaret.time_series import TSForecastingExperiment
```

```python
exp = TSForecastingExperiment()
exp.setup(data=train_data, fh=3)
```

```python
exp.models()
```

```python
model = exp.create_model("arima")
```

```python
# Out-of-sample Forecasts
y_predict = exp.predict_model(model)
y_predict
```

```python
exp.plot_model(estimator=model)
```

```python
exp.plot_model(estimator=model, data_kwargs={'fh': 36})
```

```python
best_baseline_models = exp.compare_models(n_select=3)
best_baseline_models
```

```python
best_tuned_models = [exp.tune_model(model) for model in best_baseline_models]
best_tuned_models
```

```python
compare_metrics = exp.pull()
```

```python
# Get model weights to use
top_model_metrics = compare_metrics.iloc[0:3]['MAE']
display(top_model_metrics)

top_model_weights = 1 - top_model_metrics/top_model_metrics.sum()
display(top_model_weights)
```

```python
blender = exp.blend_models(best_tuned_models, method='mean', weights=top_model_weights.values.tolist())
```

```python
y_predict = exp.predict_model(blender)
print(y_predict)
exp.plot_model(estimator=blender)
```

```python
exp.plot_model(estimator=blender, data_kwargs={'fh': 36})
```
