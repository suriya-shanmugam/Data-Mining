```python
directory = '/content/restaurant_dataset/'
```

```python
import pandas as pd
import numpy as np
```

```python
train_data = pd.read_csv(directory+'RestaurantMetadata.csv')
```

```python
train_data.head()
```

```python
from pycaret.clustering import *
s = setup(train_data, session_id = 123)
```

```python
kmeans = create_model('kmeans')
```

```python
models()
```

```python
kmeans_cluster = assign_model(kmeans)
kmeans_cluster
```

```python
plot_model(kmeans, plot = 'cluster')
```

```python
Affinity_Propagation = create_model('ap')
```

```python
affinity_cluster = assign_model(Affinity_Propagation)
affinity_cluster
```

```python
plot_model(Affinity_Propagation, plot = 'cluster')
```
