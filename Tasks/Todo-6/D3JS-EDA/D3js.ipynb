```python
import pandas as pd
df = pd.read_csv('./sample_data/dog_breeds.csv')
df.head()
```

```python
# List of names to filter
names_to_filter = ['Golden Retriever', 'Dachshund']
filtered_df = df[df['Name'].isin(names_to_filter)]
#print(filtered_df)
data = filtered_df.to_json(orient='records')
print(data)
```

```python
from IPython.display import display, HTML

display(HTML('''

<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Life Expectancy Comparison</title>
    <script src="https://d3js.org/d3.v6.min.js"></script>
    <style>
        .bar {
            fill-opacity: 0.7;
        }

        .bar:hover {
            fill-opacity: 1;
        }

        .axis text {
            font-size: 12px;
        }

        .axis path,
        .axis line {
            fill: none;
            stroke: #000;
            shape-rendering: crispEdges;
        }
    </style>
</head>
<body>
    <h1> Life Expectancy Comparison Bar Chart - Golden vs Dachshund </h1>
    <svg width="5000" height="5000"></svg>

    <script>
        // Sample Data
        const data = ''' + data + ''';

        // Set dimensions and margins
        const margin = { top: 40, right: 30, bottom: 50, left: 60 };
        const width = 800 - margin.left - margin.right;
        const height = 500 - margin.top - margin.bottom;

        // Append SVG object to body
        const svg = d3.select("svg")
            .attr("width", width + margin.left + margin.right)
            .attr("height", height + margin.top + margin.bottom)
            .append("g")
            .attr("transform", `translate(${margin.left},${margin.top})`);

        // List of subgroups (min and max life expectancy)
        const subgroups = ["min_life_expectancy", "max_life_expectancy"];

        // List of countries (groups)
        const groups = data.map(d => d.Name);

        // X scale for groups (countries)
        const x0 = d3.scaleBand()
            .domain(groups)
            .range([0, width])
            .padding(0.2);

        // X scale for subgroups
        const x1 = d3.scaleBand()
            .domain(subgroups)
            .range([0, x0.bandwidth()])
            .padding(0.05);

        // Y scale
        const y = d3.scaleLinear()
            .domain([0, d3.max(data, d => Math.max(d.min_life_expectancy, d.max_life_expectancy))])
            .nice()
            .range([height, 0]);

        // Color scale
        const color = d3.scaleOrdinal()
            .domain(subgroups)
            .range(['#1f77b4', '#ff7f0e']); // Different colors for min and max

        // Add X axis
        svg.append("g")
            .attr("transform", `translate(0,${height})`)
            .call(d3.axisBottom(x0));




        // Add Y axis
        svg.append("g")
            .call(d3.axisLeft(y));

        // Draw the bars
        svg.append("g")
            .selectAll("g")
            .data(data)
            .join("g")
            .attr("transform", d => `translate(${x0(d.Name)},0)`)
            .selectAll("rect")
            .data(d => subgroups.map(key => ({ key: key, value: d[key] })))
            .join("rect")
            .attr("x", d => x1(d.key))
            .attr("y", d => y(d.value))
            .attr("width", x1.bandwidth())
            .attr("height", d => height - y(d.value))
            .attr("fill", d => color(d.key))
            .attr("class", "bar");

        // Add legend
        const legend = svg.selectAll(".legend")
            .data(subgroups)
            .enter().append("g")
            .attr("class", "legend")
            .attr("transform", (d, i) => `translate(0,${i * 20})`);

        legend.append("rect")
            .attr("x", width - 18)
            .attr("width", 18)
            .attr("height", 18)
            .style("fill", color);

        legend.append("text")
            .attr("x", width - 24)
            .attr("y", 9)
            .attr("dy", ".35em")
            .style("text-anchor", "end")
            .text(d => d);

    </script>
</body>
</html>


'''))
```

```python
from IPython.display import display, HTML

display(HTML('''
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Life Expectancy Comparison Line Chart</title>
    <script src="https://d3js.org/d3.v6.min.js"></script>
    <style>
        .line {
            fill: none;
            stroke-width: 2;
        }

        .line-min {
            stroke: steelblue;
        }

        .line-max {
            stroke: orange;
        }

        .axis text {
            font-size: 12px;
        }

        .axis path,
        .axis line {
            fill: none;
            stroke: #000;
            shape-rendering: crispEdges;
        }

        .dot {
            fill: steelblue;
            stroke: black;
            stroke-width: 1px;
        }

        .dot-max {
            fill: orange;
        }
    </style>
</head>
<body>
  <h1> Male Vs Female - Life Expectancy Comparison Line Chart </h1>
    <svg width="800" height="500"></svg>

    <script>
        // Sample Data
        const data = ''' + data + ''';

        // Set dimensions and margins
        const margin = { top: 40, right: 30, bottom: 80, left: 60 }; // Increased bottom margin for vertical text
        const width = 800 - margin.left - margin.right;
        const height = 500 - margin.top - margin.bottom;

        // Append SVG object to body
        const svg = d3.select("svg")
            .attr("width", width + margin.left + margin.right)
            .attr("height", height + margin.top + margin.bottom)
            .append("g")
            .attr("transform", `translate(${margin.left},${margin.top})`);

        // X scale for countries (ordinal scale for names)
        const x = d3.scaleBand()
            .domain(data.map(d => d.Name))
            .range([0, width])
            .padding(0.1);

        // Y scale for life expectancy
        const y = d3.scaleLinear()
            .domain([0, d3.max(data, d => Math.max(d.max_height_male, d.max_height_female))])
            .nice()
            .range([height, 0]);

        // Add X axis with vertical labels
        const xAxis = svg.append("g")
            .attr("transform", `translate(0,${height})`)
            .call(d3.axisBottom(x));

        xAxis.selectAll("text")
            .attr("transform", "rotate(-45)") // Rotate text
            .attr("dy", "0.35em") // Vertical adjustment
            .attr("dx", "-0.5em") // Horizontal adjustment
            .style("text-anchor", "end"); // Anchor text

        // Add Y axis
        svg.append("g")
            .call(d3.axisLeft(y));

        // Line generator for min life expectancy
        const lineMin = d3.line()
            .x((d) => x(d.Name) + x.bandwidth() / 2)  // Center the points
            .y((d) => y(d.max_height_female));

        // Line generator for max life expectancy
        const lineMax = d3.line()
            .x((d) => x(d.Name) + x.bandwidth() / 2)  // Center the points
            .y((d) => y(d.max_height_male));

        // Add the min life expectancy line
        svg.append("path")
            .datum(data)
            .attr("class", "line line-min")
            .attr("d", lineMin);

        // Add the max life expectancy line
        svg.append("path")
            .datum(data)
            .attr("class", "line line-max")
            .attr("d", lineMax);

        // Add dots for min life expectancy
        svg.selectAll(".dot-min")
            .data(data)
            .join("circle")
            .attr("class", "dot")
            .attr("cx", d => x(d.Name) + x.bandwidth() / 2)
            .attr("cy", d => y(d.max_height_female))
            .attr("r", 5);

        // Add dots for max life expectancy
        svg.selectAll(".dot-max")
            .data(data)
            .join("circle")
            .attr("class", "dot dot-max")
            .attr("cx", d => x(d.Name) + x.bandwidth() / 2)
            .attr("cy", d => y(d.max_height_male))
            .attr("r", 5);

        // Add labels for the min life expectancy dots
        svg.selectAll(".text-min")
            .data(data)
            .join("text")
            .attr("x", d => x(d.Name) + x.bandwidth() / 2)
            .attr("y", d => y(d.max_height_female) - 10) // Position above the dot
            .attr("text-anchor", "middle")
            .text(d => d.max_height_female);

        // Add labels for the max life expectancy dots
        svg.selectAll(".text-max")
            .data(data)
            .join("text")
            .attr("x", d => x(d.Name) + x.bandwidth() / 2)
            .attr("y", d => y(d.max_height_male) - 10) // Position above the dot
            .attr("text-anchor", "middle")
            .text(d => d.max_height_male);

    </script>
</body>
</html>
'''))
```
