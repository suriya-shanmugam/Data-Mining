```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_swiss_roll
from sklearn.manifold import LocallyLinearEmbedding, TSNE, Isomap, MDS
from sklearn.decomposition import PCA, KernelPCA, IncrementalPCA, FactorAnalysis
from umap import UMAP
import plotly.express as px
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python


from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Generate sample dataset
data, _ = make_swiss_roll(n_samples=1000, noise=0.1)
df = pd.DataFrame(data, columns=["x", "y", "z"])
```

1. Locally Linear Embedding (LLE)
python


```python
lle = LocallyLinearEmbedding(n_neighbors=12, n_components=2)
lle_data = lle.fit_transform(data)
px.scatter(x=lle_data[:, 0], y=lle_data[:, 1], title="LLE").show()
```

t-SNE

```python
tsne = TSNE(n_components=2, random_state=42, perplexity=30)
tsne_data = tsne.fit_transform(data)
px.scatter(x=tsne_data[:, 0], y=tsne_data[:, 1], title="t-SNE").show()
```

3. ISOMAP



```python
isomap = Isomap(n_neighbors=12, n_components=2)
isomap_data = isomap.fit_transform(data)
px.scatter(x=isomap_data[:, 0], y=isomap_data[:, 1], title="ISOMAP").show()
```

4. UMAP with Interactive Visualization


```python
umap = UMAP(n_neighbors=15, n_components=2, random_state=42)
umap_data = umap.fit_transform(data)
px.scatter(x=umap_data[:, 0], y=umap_data[:, 1], title="UMAP").show()
```

5. MDS

```python
mds = MDS(n_components=2, random_state=42)
mds_data = mds.fit_transform(data)
px.scatter(x=mds_data[:, 0], y=mds_data[:, 1], title="MDS").show()
```

6. Randomized PCA

```python
random_pca = PCA(n_components=2, svd_solver='randomized', random_state=42)
random_pca_data = random_pca.fit_transform(data)
px.scatter(x=random_pca_data[:, 0], y=random_pca_data[:, 1], title="Randomized PCA").show()
```

7. Kernel PCA


```python
kernel_pca = KernelPCA(n_components=2, kernel='rbf')
kernel_pca_data = kernel_pca.fit_transform(data)
px.scatter(x=kernel_pca_data[:, 0], y=kernel_pca_data[:, 1], title="Kernel PCA").show()
```

8. Incremental PCA

```python
incremental_pca = IncrementalPCA(n_components=2)
incremental_pca_data = incremental_pca.fit_transform(data)
px.scatter(x=incremental_pca_data[:, 0], y=incremental_pca_data[:, 1], title="Incremental PCA").show()
```

9. Factor Analysis


```python
factor_analysis = FactorAnalysis(n_components=2, random_state=42)
factor_analysis_data = factor_analysis.fit_transform(data)
px.scatter(x=factor_analysis_data[:, 0], y=factor_analysis_data[:, 1], title="Factor Analysis").show()
```

10. Autoencoders


```python
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.models import Model

# Normalize the data
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data)

# Define Autoencoder architecture using functional API
input_layer = Input(shape=(scaled_data.shape[1],))
hidden_layer_1 = Dense(64, activation='relu')(input_layer)
bottleneck = Dense(2, activation='relu', name="bottleneck")(hidden_layer_1)
hidden_layer_2 = Dense(64, activation='relu')(bottleneck)
output_layer = Dense(scaled_data.shape[1], activation='sigmoid')(hidden_layer_2)

# Define Autoencoder and Encoder models
autoencoder = Model(inputs=input_layer, outputs=output_layer)
encoder = Model(inputs=input_layer, outputs=bottleneck)  # Encoder extracts the bottleneck layer's output

# Compile and train the Autoencoder
autoencoder.compile(optimizer='adam', loss='mse')
autoencoder.fit(scaled_data, scaled_data, epochs=50, batch_size=32, verbose=0)

# Use the Encoder to reduce dimensions
encoded_data = encoder.predict(scaled_data)

# Visualize the reduced data
import plotly.express as px
px.scatter(x=encoded_data[:, 0], y=encoded_data[:, 1], title="Autoencoder Dimensionality Reduction").show()
```
